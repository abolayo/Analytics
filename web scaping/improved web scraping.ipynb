```python
url ='https://www.airbnb.com/s/Mayrhofen--Austria/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&date_picker_type=calendar&query=Mayrhofen%2C%20Austria&place_id=ChIJbzLYLzjdd0cRDtGuTzM_vt4&checkin=2021-02-06&checkout=2021-02-13&adults=4&source=structured_search_input_header&search_type=autocomplete_click'
def get_listings(search_page):
    soup = BeautifulSoup(requests.get(search_page).content, 'html.parser')
    listings = soup.find_all('div', 'c1l1h97y')

    return listings
```

```python
# checking.... count the numbers of urls
len(get_listings(url))
```

```python
# let's try next page, counts of urls( data variable on each web page)
new_url = url + '&items_offset=20'
len(get_listings(new_url))
```

```python
# let's iterate through all 15 pages(of intrest)
all_listings = []
for i in range(15):
    offset = 20 * i
    new_url = url + f'&items_offset={offset}'
    new_listings = get_listings(new_url)
    all_listings.extend(new_listings)
    
    # let's check if it's scraping
    print(len(all_listings))
```

```python
# 1. build all urls
def build_urls(main_url, listings_per_page=20, pages_per_location=15):
    url_list = []
    for i in range(pages_per_location):
        offset = listings_per_page * i
        url_pagination = main_url + f'&items_offset={offset}'
        url_list.append(url_pagination)
    
    return url_list
```

```python
# build a list of URLs
url_list = build_urls(url)
url_list
```

```python
# First Generation :)
#Extract features of interest
def extract_basic_features(listings_html):
    features_dict = {}
    
    url = listings_html.find('a').get('href')
    name = listings_html.find("div", {"class": "t1jojoys"}).get_text()
    beds = listings_html.find('span',{'class': 'dir dir-ltr'}).get_text()
    price = listings_html.find('div',{'class': '_1jo4hgw'}).get_text()
    description = listings_html.find("span", {"class": "t6mzqp7"}).get_text()
    
    features_dict['url'] = url
    features_dict['name'] = name
    features_dict['beds'] =beds
    features_dict['price'] = price
    features_dict['description'] = description
    
    return features_dict
```

```python
# 2. Iteratively scrape pages
def process_search_pages(url_list):
    features_list = []
    for page in url_list:
        listings = get_listings(page)
        for listing in listings:
            features = extract_basic_features(listing)
            features_list.append(features)

    return features_list
```

```python
# try for one page
base_features = process_search_pages(url_list)
```

```python
df = pd.DataFrame.from_dict(base_features)
df.head()
```
